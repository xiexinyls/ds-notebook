# Inferential Statistical Analysis

There are two main branches for inferential statistics analysis: **explanatory modeling** and **predictive modeling**.

The major job for explanatory modeling is to find the relationship between data. (from SAS material)

* How is X related to Y?
* Sample size are small.
* Include few variables.
* Use p-values and confidence intervals.

The major job for predictive modeling is to find the relationship between data. (from SAS material)

* If you know X, can you predict Y?
* Sample size are large.
* Include many variables.
* Need holdout verification.

### Notations and Terminology

Usually, population parameters (or distribution population) are represented by Greek letters while sample statistics are represented by English letters.

Independent variables are also called input variable, control variable, explanatory variable, and predictor variable. Dependent variable is also called response variable, outcome variable, and target variable.

Quantitative data are represented by numbers, which can be continuous or discrete. Arithmetic operations can be applied to quantitative data. Categorical data are represented by labels, which can be nominal (no ordering, classification) or ordinal (with ordering). It has no meaning to apply arithmetic operations to categorical data.

Before You Do Analysis, check for outliers, data entry errors. Check range, frequency, shape of data, and uniqueness of data.

### Basic Statistics

For quantitative data, we can calculate basic statistics such as mean, median, mode, frequency distribution, percentile, range, coefficient variation to see what the data look like. SAS provides `means` procedure to get basics statistics. By default, it first classifies data into groups according to `class`, finds all the quantitative variables and calculates the mean, min, max, and standard deviation. `maxdec` is maximum decimal digit number. `fw` means width. `printalltypes` outputs the all type results even with `class` specified.

	proc means data=input [maxdec=2] [fw=10] [printalltypes];
	  class var1;
	  var var2;
	run;

You may specify the statistics you need.

	proc means data=input mean median min max range qrange stderr clm alpha=0.01;
	  class var1;
	  var var2;
	run;

### Frequency table and Histogram

One way

	proc freq data=input;
	  table var1;
	run;

Multiple way

	proc freq data=input;
	  table var1*var2;
	run;


To remove cumulative part, add `nocum` option

	proc freq data=input;
	  table var1 /nocum;
	run;

### A Single `univariate` Procedure

Histogram is another way to look at frequency distribution visually. SAS provides a simple `univariate` procedure to check all the aspects of basic statistics

	proc univariate data=statdata.testscores;
	  id varid;
	  var var1;
	  hist var1 / normal(mu=est sigma=est );
	  inset skew kurt / position=ne;
	  probplot var1 / normal(mu=est sigma=est );
	  inset skew kurt;
	run;

### Random Data Sampling

The SAS procedure `surveyselect` can randomly sample data. To do sampling without replacement, use

	proc surveyselect data=input out=output method=srs seed=idseed sampsize=100;
	run;

To do sampling with replacement, use

	proc surveyselect data=input out=output method=urs seed=idseed sampsize=100;
	run;

To sample only a percentage (80% for example), use

	proc surveyselect data=input out=output method=srs seed=idseed samprate=0.8;
	run;
